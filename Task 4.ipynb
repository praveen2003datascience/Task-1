```python
import pandas as pd
import seaborn as sns
```

```python
data = pd.read_csv(r"C:\Users\Praveen T\Downloads\Twitter_cleaned.csv")
```

```python
# Analyze the distribution of sentiment
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(10, 6))
sns.countplot(data=data, x='Sentiment')
plt.title('Distribution of Sentiment')
plt.xlabel('Sentiment')
plt.ylabel('Count')
plt.show()
```

```python
# Calculate the frequency of each topic
```

```python
topic_columns = [col for col in data.columns if col not in ['Unnamed: 0', 'Sentiment', 'Tweet_word_count', 'Tweet_char_count', 'Tweet_clean']]

topic_frequencies = data[topic_columns].sum().sort_values(ascending=False)

# Display the top topics
print(topic_frequencies)
```

```python
# Plot the frequency of mentions for each topic
plt.figure(figsize=(12, 8))
sns.barplot(x=topic_frequencies.index, y=topic_frequencies.values, palette='viridis')
plt.title('Frequency of Mentions for Each Topic')
plt.xlabel('Topic')
plt.ylabel('Frequency')
plt.xticks(rotation=90)
plt.show()
```

```python
# Sentiment vs. Tweet word count
plt.figure(figsize=(10, 6))
sns.boxplot(data=data, x='Sentiment', y='Tweet_word_count')
plt.title('Tweet Word Count by Sentiment')
plt.xlabel('Sentiment')
plt.ylabel('Tweet Word Count')
plt.show()
```

```python
# Sentiment vs. Tweet character count
plt.figure(figsize=(10, 6))
sns.boxplot(data=data, x='Sentiment', y='Tweet_char_count')
plt.title('Tweet Character Count by Sentiment')
plt.xlabel('Sentiment')
plt.ylabel('Tweet Character Count')
plt.show()
```

```python
# Example: Sentiment distribution for different topics/brands
topics = ['ApexLegends', 'AssassinsCreed', 'Borderlands', 'CS-GO', 'Overwatch', 'PlayStation5(PS5)', 'Xbox(Xseries)']

for topic in topics:
    plt.figure(figsize=(10, 6))
    sns.countplot(data=data, x='Sentiment', hue=topic)
    plt.title(f'Sentiment Distribution for {topic}')
    plt.xlabel('Sentiment')
    plt.ylabel('Count')
    plt.legend(title=topic)
    plt.show()
```
